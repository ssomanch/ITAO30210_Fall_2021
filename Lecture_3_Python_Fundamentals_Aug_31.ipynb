## Conditional Execution: if statement

Note the indentation after the if statement

## Conditional Execution: if else statement

```python
name = input('Enter a name')
if name == 'Alice':
    print('Hi, Alice!')
else:
    print("No! You aren't Alice!")
```

## Conditional Execution: if elif else statement

```python
name = input('Enter a name')
if name == 'Alice':
    print('Hi, Alice!')
elif name =='Sam':
    print('Hi, Sam')
else:
    print("No! You aren't Alice or Sam!")
```

## Fancy print statements

```python
age = 23
height = 5.23623
print("The age is {0} and the height is {1:.2f}".format(age, height))
```

```python
age = 23
height = 5.23623
print("The age is {1} and the height is {0:.3f}".format(height, age))
```

```python
age = 23
height = 5.23623
print(f"The age is {age} and the height is {height:.2f}")
```
