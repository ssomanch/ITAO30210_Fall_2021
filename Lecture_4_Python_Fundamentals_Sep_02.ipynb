## Practice #4: Keep guessing!

Write a program that allows the user to guess the answer to the question, “Which U.S. president was born on July 4?”

Sample Interaction

Which US president was born on July 4th?: Barack Obama

Enter another guess: Donald Trump

Enter another guess: George Bush

Enter another guess: Calvin Coolidge

Thank you!

```python
name = input("Which U.S. president was born on July 4th? ")

# While the name is not Calvin Coolidge
# # Get more guesses from the user

while name.lower() != 'calvin coolidge':
    name = input("Enter another guess: ")
    
print("Thank you!")
```

## Modified Practice #4: Limited guesses with hints

Write a program that allows the user **ten tries** to answer the question, “Which U.S. president was born on July 4?”

After three incorrect guesses, the program should display the hint, “He once said, ‘If you don’t say anything, you won’t be called upon to repeat it.’”

After seven incorrect guesses, the program should give the hint, “His nickname was ‘Silent Cal.’


```python
# 1. Limit the number of guesses to 10

# 2. If the user is unable to answer after 10 guesses provide them the right answer

# 3. Provide hints after 3 and 7 wrong guesses


name = input("Which U.S. president was born on July 4th? ")
counter = 1

# While the name is not Calvin Coolidge and less than 10 guesses are made
# # Get more guesses from the user

while (name.lower() != 'calvin coolidge') and (counter<10):
    if counter == 3:
        print("He once said 'If you do not say anything, you will not be called to repeat it'")
    if counter == 7:
        print("His nickname was 'Silent Cal'")
    name = input("Enter another guess: ")
    counter = counter + 1
    
# If they did not get the right asnwer, I want to privide the answer
# Else I want to say thank you

if name.lower() != 'calvin coolidge':
    print("The correct answer is Calvin Coolidge!")
else:
    print("Thank you!")
```

### Practice #5: Write a program to accept a number n and prints the sum of first n positive integers

###### Sample Interaction:

Enter a number: 5

The sum of first 5 numbers is 15


```python
num = int(input("Enter a number: "))

sum_n = 0

for i in range(num+1):
    sum_n = sum_n + i

print(sum_n)
```
