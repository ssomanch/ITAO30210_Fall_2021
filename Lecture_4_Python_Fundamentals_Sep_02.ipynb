## Nested Conditionals

```python
a = 1000
b = 5
c = 20
if a > b:
    if a < c:
        print("The value of a is between b and c")
```

```python
a = 50
b = 5
c = 20
if a > b:
    if a < c:
        print("The value of a is between b and c")
    else:
        print("Something")
elif a < b:
    print ("The value of a is less than b")
else:
    print("The value of a is same as b")
```
